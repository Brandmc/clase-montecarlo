```python
def rollGame():
    roll=np.random.randint(1,7)
    if roll==6:
        print("El resultado fue %s, PERDISTE!!! juega otra vez" %roll)
    elif roll<=3:
        print("el resultdo esta entre 1 y 3. GANASTE!!!")
```

```python
rollGame()
```

```python
def rollGame():
    roll=np.random.randint(1,7)
    if roll>=4:
        return False
    else:
        return True
```

```python
def apostador_simple(fondos,ini_apuesta,n_apuestas):
    valor=fondos
    apuesta=ini_apuesta
    apuestaActual=0
    while apuestaActual<n_apuestas:
        if rollGame():
            valor+=apuesta
        else:
            valor-=apuesta
        apuestaActual+=1
        print("fondos: ",valor)
```

```python
apostador_simple(10000,100,5)
```

```python
def apostador_simple(fondos,ini_apuesta,n_apuestas):
    valor=fondos
    apuesta=ini_apuesta
    apuestaActual=0
    wX,wY=[],[]
    while apuestaActual<n_apuestas:
        if rollGame():
            valor+=apuesta
            wX.append(apuestaActual)
            wY.append(valor)
        else:
            valor-=apuesta
            wX.append(apuestaActual)
            wY.append(valor)
        apuestaActual+=1
    plt.plot(wX,wY,lw=0.5)
```

```python
apostador_simple(10000,100,500)
```

### cadenas e markov

```python
n=7000000000 #tamaño de la acumulacion maxima
a=0.5/n #razon de nacimientos
b=0.2/n #razon de muertes
npasos=1000
x=np.zeros(npasos)
x[0]= #poblacion inicial
for t in range(npasos-1):
    if 0<x[t]<n-1:
        nacimientos=np.random.rand()<=a*x[t]
        muerte=np.random.rand()<=a*x[t]
        x[t+1]=x[t]+nacimientos*1-muerte*1
    else:
        x[t+1]=x[t]
```

```python
plt.plot(x,c="orange")
plt.show()
```
